```python
df["cleaned_text"] = df["text"]
```

```python
df["cleaned_text"] = df["cleaned_text"].str.lower()
```

```python
df["cleaned_text"] = df["cleaned_text"].str.replace(r"\r\n", " ", regex=True)
```

```python
df["cleaned_text"] = df["cleaned_text"].str.replace(r"\s+", " ", regex=True).str.strip()
```
